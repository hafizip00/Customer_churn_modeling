```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(1234)
num_obs = 100
x1 = np.random.uniform(1,10, num_obs)
x2 = np.random.uniform(1,10,num_obs)
const = np.ones(num_obs)
epa = np.random.normal(0,.5,num_obs)

b = 1.5
theta_01 = 2
theta_02 = 5

y = b*const + x1*theta_01 + x2*theta_02 + epa

x_mat = np.array([const, x1,x2])
```

```python
x_mat = np.array([const, x1,x2])
```

```python
x_mat.T
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression(fit_intercept=False)
lr_model.fit(x_mat.T, y)
lr_model.coef_
```

```python
# Using Linear Algebra
np.linalg.inv(np.dot(x_mat.T,x_mat)).dot(x_mat.T).dot(y)
```
